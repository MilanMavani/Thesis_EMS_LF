# Current ML Workflow

## Goal
The aim of this workflow is to build a structured and reproducible pipeline for load forecasting.

## Main Flow
**Raw data → Concatenation → Cleaning → Resampling → Feature Engineering → Spliting → Model Training → Tracking → Stored Outputs**

---

## 1. Concatenation
Different raw files are combined into one time-indexed dataset.


To create one continuous dataset for further processing.

---

## 2. Cleaning
The data is cleaned by handling:
- wrong timestamps
- duplicates
- missing values
- sensor errors / outliers


To improve data quality before model training.

---

## 3. Resampling
The original data is converted to the required resolution such as:
- 15-minute
- e.g. 30-minute

To avoide spikes from 5 min reso
---

## 4. Feature Engineering
New input features are added, such as:
- calendar features
- lag features
- rolling mean / std features

 
These features help the model learn load patterns better.

---

## 5. Splitting
The dataset is split into:
- training set 70 %
- validation set 15 %
- test set 15 %

A **Chronological split** is used.

 
Because forecasting must be in time order and avoid leakage. underfit and overfit 

---

## 6. Model Training
Different ML models are trained and compared on the same dataset.

Examples:
- Linear Regression
- Ridge
- Random Forest
- XGBoost(nan friedly)
- CatBoost (nan friedly)
- HistGradientBoosting (nan friedly)




---

## 7. Tracking
Each experiment is tracked by saving:
- run ID
- model name
- metrics
- predictions
- trained model
- plots



To make the thesis work organized and reproducible.

---

## 8. Outputs Stored
The workflow stores:
- prepared datasets
- trained models
- predictions
- error metrics
- plots
- experiment logs
- feature importance

---

This workflow helps me perform experiments in a **modular, clear, reproducible and Efficient** way. so far

# Forecasting Approach: Current Work and Next Step

## 1. Current Work: Point Forecasting
At the moment, my current experiments are based on **point forecasting**.

### What it means
For each input sample, the model predicts **one future load value**.

### Example
- predict the next single load value
- or predict one value at a specific future time step

### Important note
Even if the prediction plot shows many timestamps over several days, it can still be **point forecasting** because the model is producing **one value at a time** for each sample.

---

## 2. Why it is called Point Forecasting
The difference is based on the **model output per sample**:

- **Point forecasting** → one future value
- **Profile forecasting** → a full future sequence / curve

So my current model output is a **single value**, therefore it is point forecasting.

---

## 3. Next Step
The next step of my thesis is to move toward a more practical forecasting setup.

I plan to extend the work in one of these directions:

- **Point forecasting with longer horizons**
  - 1-day-ahead
  - 1-week-ahead

and/or

- **Profile forecasting**
  - predict the full future load profile
  - for example, the complete next-day load curve

---

## 4. Profile Forecasting
In profile forecasting, the model predicts **multiple future values together**.

### Example
Instead of predicting only one point, the model predicts:
- the full next-day profile
- or the full next-week profile

This is more useful for energy management and planning.

---

## 5. Why this extension is important
Point forecasting is useful as a baseline and for model comparison.

But for practical EMS use, it is often more useful to know:
- the whole next-day demand pattern
- expected peaks
- the overall future load shape

That is why profile forecasting is an important next step.

---

for:
