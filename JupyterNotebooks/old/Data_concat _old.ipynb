```python
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
pio.renderers.default = "notebook_connected"

def plot_feature(df, cols=None, start=None, end=None, connectgaps=False):
    d = df.copy()

    # ensure datetime index
    d.index = pd.to_datetime(d.index, errors="coerce")
    d = d[~d.index.isna()].sort_index()

    # time filter
    if start is not None or end is not None:
        d = d.loc[start:end]

    # choose columns
    if cols is None:
        cols = d.columns.tolist()

    # numeric convert (important if some columns are strings)
    d[cols] = d[cols].apply(pd.to_numeric, errors="coerce")

    # plot multiple columns
    fig = px.line(d, x=d.index, y=cols,)
    fig.update_traces(connectgaps=connectgaps)
    fig.show()
```

## Data Concatenate

### Concatenate all features in to single Dataset 
        Outer join

```python
ROOT = Path(r"C:\Data_analysis\Thesis\Data\01_Raw_Data")  # main folder containing all signal subfolders
SAVE_PER_SIGNAL = False  # True if per-signal combined files needed

# Helper Functionf

def find_header_row(path: Path) -> int:

    with open(path, "r", encoding="utf-8", errors="replace") as f:
        for i, line in enumerate(f):
            if line.strip().startswith("Time"):
                return i
    raise ValueError(f"Header line starting with 'Time' not found in: {path}")


def parse_time_mixed(s: pd.Series) -> pd.Series:

    s = s.astype(str).str.strip()

    t1 = pd.to_datetime(s, format="%m/%d/%Y %I:%M:%S %p", errors="coerce")  # AM/PM + seconds
    t2 = pd.to_datetime(s, format="%m/%d/%Y %H:%M", errors="coerce")        # 24h no seconds
    t3 = pd.to_datetime(s, format="%m/%d/%Y %H:%M:%S", errors="coerce")     # 24h + seconds
    t4 = pd.to_datetime(s, format="%Y-%m-%d %H:%M:%S", errors="coerce")     # ISO + seconds
    t5 = pd.to_datetime(s, format="%Y-%m-%d %H:%M", errors="coerce")        # ISO no seconds

    out = t1.fillna(t2).fillna(t3).fillna(t4).fillna(t5)

    # last fallback for any remaining odd cases
    if out.isna().any():
        out = out.fillna(pd.to_datetime(s, errors="coerce"))

    return out


def robust_read_csv_first_two_cols(csv_path: Path, sep: str = ",") -> pd.DataFrame:

    header_row = find_header_row(csv_path)

    # Prefer C engine (fast and avoids python field limit issues)
    try:
        df = pd.read_csv(
            csv_path,
            sep=sep,
            header=header_row,
            engine="c",
            usecols=[0, 1],  # ignore trailing empty column caused by trailing commas
            encoding="utf-8",
        )
        return df
    except Exception:
        # Fallback: python engine + increase field limit
        csv.field_size_limit(min(sys.maxsize, 10_000_000))
        df = pd.read_csv(
            csv_path,
            sep=sep,
            header=header_row,
            engine="python",
            usecols=[0, 1],
            encoding="utf-8",
            on_bad_lines="skip",  # skip truly broken lines (optional)
        )
        return df

# Pipeline
def read_year(signal_folder: Path, year: str) -> tuple[pd.DataFrame, str]:
    csv_path = signal_folder / f"{year}.csv"
    df = robust_read_csv_first_two_cols(csv_path, sep=",")

    df.columns = [c.strip() for c in df.columns]

    # original names from file
    time_col = df.columns[0]
    sig_col = df.columns[1]  # keep csv header name as feature name

    # rename time to a standard key
    df = df.rename(columns={time_col: "Time"})

    # robust parse time (handles BOTH formats)
    df["Time"] = parse_time_mixed(df["Time"])

    # numeric signal
    df[sig_col] = pd.to_numeric(df[sig_col], errors="coerce")

    # clean + deduplicate
    df = df.dropna(subset=["Time"]).sort_values("Time").drop_duplicates("Time", keep="last")

    return df[["Time", sig_col]], sig_col


def read_signal(signal_folder: Path) -> tuple[pd.DataFrame, str] | None:
    dfs = []
    sig_name = None

    for year in ["2025", "2026"]:
        csv_path = signal_folder / f"{year}.csv"
        if csv_path.exists():
            df_year, this_sig_name = read_year(signal_folder, year)

            # handle rare case: different header name across years
            if sig_name is None:
                sig_name = this_sig_name
            elif sig_name != this_sig_name:
                df_year = df_year.rename(columns={this_sig_name: sig_name})

            dfs.append(df_year)

    if not dfs:
        return None

    df_all = pd.concat(dfs, ignore_index=True).sort_values("Time")
    df_all = df_all.drop_duplicates("Time", keep="last")

    if SAVE_PER_SIGNAL:
        out_path = signal_folder / f"{sig_name}_2025_2026.csv"
        df_all.to_csv(out_path, index=False, encoding="utf-8")
        print("Saved per-signal:", out_path)

    return df_all, sig_name


# ---- main loop to combine all signals ----
signal_folders = sorted([p for p in ROOT.iterdir() if p.is_dir()], key=lambda p: p.name.lower())

combined = None
used_names = set()
skipped = []

for i, folder in enumerate(signal_folders, start=1):
    try:
        res = read_signal(folder)
        if res is None:
            skipped.append(folder.name)
            continue

        df_sig, sig_name = res

        # Ensure unique column names across all signals
        final_name = sig_name
        if final_name in used_names:
            final_name = f"{sig_name}__{folder.name}"  # disambiguate
            df_sig = df_sig.rename(columns={sig_name: final_name})

        used_names.add(final_name)

        if combined is None:
            combined = df_sig
        else:
            combined = combined.merge(df_sig, on="Time", how="outer")

        print(f"Processed {i}/{len(signal_folders)}: {folder.name} -> {final_name}")

    except Exception as e:
        skipped.append(folder.name)
        print(f"Skipped {folder.name} because: {e}")

if combined is None:
    raise RuntimeError("No signal folders were successfully processed.")

combined = combined.sort_values("Time")

# Save csv 
out_path = Path(r"C:\Data_analysis\Thesis\Data\02_Preprocessing\LF_data_Raw.csv") 
combined.to_csv(out_path, index=False, encoding="utf-8")

# print status of execution
print("\nDONE :)")
print("Saved big file:", out_path)
print("Final shape:", combined.shape)
if skipped:
    print("Skipped folders:", skipped)
```
