```python
from pathlib import Path
import sys
import pandas as pd

PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

from src.config import EXPERIMENT_LOG_FILE

log_df = pd.read_csv(EXPERIMENT_LOG_FILE)

metric_cols = ["test_MAE", "test_RMSE", "test_MAPE", "test_sMAPE", "test_R2"]
for col in metric_cols:
    log_df[col] = pd.to_numeric(log_df[col], errors="coerce")

log_df.head()
```

```python
academy_results = log_df[
    (log_df["target"] == "BU_TotActPwr_Tech_Room") &
    (log_df["feature_set_name"] == "all_features_v1")
].copy()  #    (log_df["target"] == "BU_TotActPwr_Tech_Room") &

academy_results = academy_results.sort_values("test_RMSE", ascending=True)

academy_results[
    [
        "run_id",
        "model_name",
        "target",
        "dataset_name",
        "data_mode",
        "frequency",
        "feature_set_name",
        "test_MAE",
        "test_RMSE",
        "test_MAPE",
        "test_sMAPE",
        "test_R2",
    ]
]
```

```python
academy_compare = log_df[
    (log_df["feature_set_name"] == "all_features_v1")
].copy()   #(log_df["target"] == "BU_TotActPwr_Academy") &

academy_compare["test_RMSE"] = pd.to_numeric(academy_compare["test_RMSE"], errors="coerce")
academy_compare = academy_compare.dropna(subset=["test_RMSE"])

best_per_dataset = (
    academy_compare.sort_values("test_RMSE", ascending=True)
    .groupby("dataset_name", as_index=False)
    .first()
)

best_per_dataset[
    [
        "dataset_name",
        "model_name",
        "target",
        "data_mode",
        "test_MAE",
        "test_RMSE",
        "test_MAPE",
        "test_sMAPE",
        "test_R2",
    ]
]
```
